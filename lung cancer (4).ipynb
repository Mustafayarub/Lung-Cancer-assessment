```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=3,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
prediction
```

```python
from sklearn import metrics
```

```python
accuracy = metrics.accuracy_score(y_test,prediction)
accuracy
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='binary')
recall = recall_score(y_test, y_pred, average='binary')
f1 = f1_score(y_test, y_pred, average='binary')
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))
```

```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

sns.set(style="whitegrid")

class LungCancerGUI:
    def __init__(self, root):
        self.root = root
        self.root.title("Lung Cancer Data Explorer & Predictor")
        self.df = None
        self.model = None
        self.label_encoders = {}

        self.create_widgets()

    def create_widgets(self):
        frm_top = ttk.Frame(self.root, padding=8)
        frm_top.pack(fill="x")

        ttk.Button(frm_top, text="Open CSV", command=self.load_csv).grid(row=0, column=0, padx=4, pady=4)
        ttk.Button(frm_top, text="Show Summary", command=self.show_summary).grid(row=0, column=1, padx=4, pady=4)

        ttk.Label(frm_top, text="Select plot:").grid(row=0, column=2, padx=(16,4))
        self.plot_choices = ["Gender distribution", "Age distribution", "Smoking vs Lung Cancer",
                             "Lung Cancer Count", "Smokers Count"]
        self.plot_var = tk.StringVar(value=self.plot_choices[0])
        self.cmb_plot = ttk.Combobox(frm_top, values=self.plot_choices, textvariable=self.plot_var, state="readonly", width=35)
        self.cmb_plot.grid(row=0, column=3, padx=4, pady=4)
        ttk.Button(frm_top, text="Plot", command=self.draw_plot).grid(row=0, column=4, padx=4, pady=4)

        ttk.Button(frm_top, text="Predict Patient", command=self.open_predict_window).grid(row=0, column=5, padx=4, pady=4)

        frm_left = ttk.Frame(self.root, padding=8)
        frm_left.pack(side="left", fill="y")

        ttk.Label(frm_left, text="Output:").pack(anchor="nw")
        self.txt = tk.Text(frm_left, width=45, height=35)
        self.txt.pack(fill="y", expand=False)

        frm_right = ttk.Frame(self.root, padding=8)
        frm_right.pack(side="left", fill="both", expand=True)
        self.fig, self.ax = plt.subplots(figsize=(6,4))
        self.canvas = FigureCanvasTkAgg(self.fig, master=frm_right)
        self.canvas.get_tk_widget().pack(fill="both", expand=True)

    def load_csv(self):
        path = filedialog.askopenfilename(filetypes=[("CSV files","*.csv"), ("All files","*.*")])
        if not path: return
        try:
            df = pd.read_csv(path)
        except Exception as e:
            messagebox.showerror("Error", f"Failed to load CSV:\n{e}")
            return
        self.df = df
        self.train_model()
        self.display_summary()

    def display_summary(self):
        self.txt.delete("1.0", tk.END)
        self.txt.insert(tk.END, f"Columns: {', '.join(self.df.columns)}\n")
        self.txt.insert(tk.END, f"Shape: {self.df.shape}\n\n")
        self.txt.insert(tk.END, "Missing values:\n")
        self.txt.insert(tk.END, f"{self.df.isnull().sum()}\n\n")
        self.txt.insert(tk.END, f"Duplicate rows: {self.df.duplicated().sum()}\n\n")
        if "LUNG_CANCER" in self.df.columns:
            self.txt.insert(tk.END, "Lung Cancer counts:\n")
            self.txt.insert(tk.END, f"{self.df['LUNG_CANCER'].value_counts()}\n\n")
        if "SMOKING" in self.df.columns:
            self.txt.insert(tk.END, "Smoking counts:\n")
            self.txt.insert(tk.END, f"{self.df['SMOKING'].value_counts()}\n\n")

    def show_summary(self):
        if self.df is None:
            messagebox.showinfo("No data", "Please load CSV first.")
            return
        self.display_summary()

    def draw_plot(self):
        if self.df is None:
            messagebox.showinfo("No data", "Please load CSV first.")
            return
        choice = self.plot_var.get()
        self.ax.clear()
        try:
            if choice == "Gender distribution":
                vc = self.df["GENDER"].value_counts()
                sns.barplot(x=vc.index, y=vc.values, palette="Set2", ax=self.ax)
                self.ax.set_title("Gender distribution")
            elif choice == "Age distribution":
                sns.histplot(self.df["AGE"], bins=15, color="skyblue", edgecolor="black", ax=self.ax)
                self.ax.set_title("Age distribution")
            elif choice == "Smoking vs Lung Cancer":
                cross = pd.crosstab(self.df["SMOKING"], self.df["LUNG_CANCER"])
                cross.plot(kind="bar", stacked=False, ax=self.ax, colormap="Set1")
                self.ax.set_title("Smoking vs Lung Cancer")
            elif choice == "Lung Cancer Count":
                vc = self.df["LUNG_CANCER"].value_counts()
                sns.barplot(x=vc.index, y=vc.values, palette="Set1", ax=self.ax)
                self.ax.set_title("Lung Cancer Count")
            elif choice == "Smokers Count":
                vc = self.df["SMOKING"].value_counts()
                sns.barplot(x=vc.index, y=vc.values, palette="Set3", ax=self.ax)
                self.ax.set_title("Smokers Count")
            self.canvas.draw()
        except Exception as e:
            messagebox.showerror("Error", f"Plotting failed: {e}")

    def train_model(self):
        df = self.df.copy()
        X = df.drop(columns=["LUNG_CANCER"])
        y = df["LUNG_CANCER"]
        for col in X.select_dtypes(include=['object']).columns:
            le = LabelEncoder()
            X[col] = le.fit_transform(X[col].astype(str))
            self.label_encoders[col] = le
        self.model = RandomForestClassifier()
        self.model.fit(X.fillna(0), y)

    def open_predict_window(self):
        if self.model is None:
            messagebox.showinfo("No model", "Load CSV first to train the model.")
            return
        win = tk.Toplevel(self.root)
        win.title("Predict Lung Cancer")
        entries = {}

        for i, col in enumerate(self.df.drop(columns=["LUNG_CANCER"]).columns):
            ttk.Label(win, text=col).grid(row=i, column=0, padx=4, pady=4)
            if col == "GENDER":
                cb = ttk.Combobox(win, values=["M","F"], state="readonly")
                cb.grid(row=i, column=1, padx=4, pady=4)
                entries[col] = cb
            elif col == "AGE":
                ent = ttk.Entry(win)
                ent.grid(row=i, column=1, padx=4, pady=4)
                entries[col] = ent
            else:
                cb = ttk.Combobox(win, values=[1,2], state="readonly")
                cb.grid(row=i, column=1, padx=4, pady=4)
                entries[col] = cb

        def predict():
            data = {}
            for col, widget in entries.items():
                val = widget.get()
                if not val:
                    messagebox.showerror("Input error", f"Please enter/select {col}")
                    return
                if col in self.label_encoders:
                    le = self.label_encoders[col]
                    val = le.transform([val])[0]
                elif col != "AGE":
                    val = int(val)
                else:
                    try:
                        val = float(val)
                    except:
                        messagebox.showerror("Input error", "AGE must be a number")
                        return
                data[col] = [val]
            df_pred = pd.DataFrame(data)
            pred = self.model.predict(df_pred)[0]
            messagebox.showinfo("Prediction Result", f"Predicted Lung Cancer: {pred}")

        ttk.Button(win, text="Predict", command=predict).grid(row=len(entries), column=0, columnspan=2, pady=8)

def main():
    root = tk.Tk()
    app = LungCancerGUI(root)
    root.geometry("1200x700")
    root.mainloop()

if __name__ == "__main__":
    main()
```
